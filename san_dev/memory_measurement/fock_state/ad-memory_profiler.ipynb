```python
def memory_plot(h_dim, mem_mean, title=None):
    x = h_dim
    y = mem_mean

    plt.scatter(x, y, color='C0')
    plt.plot(x, y, color='C0', alpha = 0.5)

    #plt.xscale("log")
    plt.xlabel("l")
    plt.ylabel("Memory usage [MiB]")
    if title is not None:
        plt.title(title)
    plt.grid()

    plt.show()
```

```python
h_dims = [int(10**i) for i in range(1, 5, 1)]
mem_ma_Nfixed = [0.5, 20.8, 2250.90, 41966.1]
# plot memory_profiler results
memory_plot(h_dims, mem_ma_Nfixed,
            title = "Fock 1 of J.C. - memory_profiler")
```
